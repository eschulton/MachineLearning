```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r-') # plot a 2d curve
```

```python
plt.hist(np.random.randn(1000, 1) * 100, 100) # histogram
```

```python
X, y = sklearn.datasets.make_circles(1000)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap="PiYG") #scatter plot
```

```python
X = np.random.randn(10, 10)
plt.imshow(X)
```

```python
X = np.linspace(1, 10, 100)
f1 = X
f2 = X**2
f3 = np.sin(X) * 50
plt.plot(X, f1)
plt.plot(X, f2)
plt.plot(X, f3)
```

```python
fig = plt.figure()
a1 = fig.add_axes([.1, .1, 1.5, 1.5])
a2 = fig.add_axes([.5, .5, .5, .5])
a1.plot(X, f1)
a2.plot(X, f3)
```
