```python
import sklearn.datasets
import sklearn.model_selection
import matplotlib.pyplot as plt
import numpy as np
import neuralnet as nn
```

```python
X, y = sklearn.datasets.make_circles(n_samples=1000, noise=0.1)
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.3)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
```

```python
X_train = X_train.T
X_test = X_test.T
y_train = y_train.T.reshape(1, -1)
y_test = y_test.T.reshape(1, -1)
```

```python
model = nn.Network(2)
model.add_layer(4, activation_name="lrelu")
model.add_layer(1, activation_name="sigmoid")
model.train(X_train, y_train)
```

```python
plt.plot(model.cost_history)
```

```python
# for visualizing the decision areas
D = np.array([[x1, x2] for x1 in np.linspace(-1.2, 1.2, 100) for x2 in np.linspace(-1.2, 1.2, 100)]).T
model.forward_propagation(D)
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(D[0,:], D[1,:], c=model.predictions, cmap='viridis', alpha=0.05)
```
